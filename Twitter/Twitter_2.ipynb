```python
from nltk.corpus import stopwords
import string
stop=stopwords.words('english')
punctuation=list(string.punctuation)
stop+=punctuation
stop
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
from nltk import pos_tag
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_reviews(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop:
            pos=pos_tag([w])
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words
```

```python
documents = [(clean_reviews(word_tokenize(tweet.values[i,5])), tweet.values[i,1]) for i in range(0,10980)]
```
